```python
cust_df.loc[cust_df['var3']==-999999,'var3'] = 2
```

```python
cust_df = cust_df.drop(columns='ID')
```

```python
cust_df
```

```python
x=cust_df[cust_df.columns.difference(["TARGET"])]
```

```python
y=cust_df['TARGET']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
x, y, test_size=0.2, random_state=42,stratify=y)
```

```python
y_train.value_counts()/y_train.count()
```

```python
y_test.value_counts()/y_test.count()
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score
```

```python
xgb_model = XGBClassifier(n_estimators=400, random_state=42)
```

```python
xgb_model.fit(x_train,y_train, early_stopping_rounds=100, eval_metric="auc",
             eval_set=[(x_train,y_train),(x_test,y_test)])
```

```python
y_pred = xgb_model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
y_pred_proba = xgb_model.predict_proba(x_test)
```

```python
roc_auc_score(y_test,y_pred_proba[:,1],average='macro')
```
